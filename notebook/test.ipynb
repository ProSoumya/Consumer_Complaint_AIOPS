# Pipeline Configuration

```python
class Config:
    def __init__(self):
        self.workdir= workdir
        self.data_dir= os.path.join(workdir,"data")
        self.csv_filename = "data.csv"
        self.csv_path = os.path.join(self.data_dir,self.csv_filename)
        self.pipeline_name ="aiops_pipeline"
        self.pipeline_root = os.path.join("aiops_pipeline_artifact")
        self.metadata_path = os.path.join("aiops_pipeline_metadata",self.pipeline_name, "metadata.db")
        
```

```python
config = Config()
```

```python
config.metadata_path
```

```python
#Pipeline configuration 
from tfx.orchestration.experimental.interactive.interactive_context import InteractiveContext
from tfx.orchestration.metadata import sqlite_metadata_connection_config
from tfx.components import CsvExampleGen
```

```python
interactive_context =InteractiveContext(
    pipeline_name = config.pipeline_name,
    pipeline_root= config.pipeline_root,
    metadata_connection_config= sqlite_metadata_connection_config(config.metadata_path)
)
```

```python
csv_example_gen = CsvExampleGen(

    input_base=config.data_dir

)
```

```python
interactive_context.run(csv_example_gen)
```
